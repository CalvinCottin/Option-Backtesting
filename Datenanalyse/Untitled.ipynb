```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
```

```python
class investor:
    def __init__(self, no, capital, entryD, asset):
        self.no = no
        self.capital = capital
        self.initial = capital
        self.entryD = entryD
        self.asset = asset
    
    def work(self, df, shareno):
        delta = timedelta(days=1)
        entry = False
        currentShares = 0
        for date in df["Date"]:
            if (date >= self.entryD):
                temp = df[df.Date == date]
                if (temp["TradingPoints"].values == "0"):
                    self.capital = self.capital - temp.Close.values*shareno
                    currentShares = currentShares + shareno
                    print("Now buying" + str(shareno) + "Shares")
                if (temp["TradingPoints"].values == "4"):
                    self.capital = self.capital + temp.Close.values*currentShares*0.5
                    currentShares = currentShares * 0.5
                    print("Now selling" + str(currentShares) + "Shares")
                if (temp["TradingPoints"].values == "5"):
                    self.capital = self.capital + temp.Close.values*currentShares
                    currentShares = 0 
                    print("Now sold all Shares")
        print((self.capital-self.initial))
            
        
```
