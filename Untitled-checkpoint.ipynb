```python
def powerset(s):
    x = len(s)
    liste = []
    for i in range(1 << x):
        liste.append(([s[j] for j in range(x) if (i & (1 << j))]))
    return liste
```

```python
def std_dev(data):
    # Get number of observations
    n = len(data)
    # Calculate mean
    mean = sum(data) / n
    # Calculate deviations from the mean
    deviations = sum([(x - mean)**2 for x in data])
    # Calculate Variance & Standard Deviation
    variance = deviations / (n - 1)
    s = variance**(1/2)
    return s


# Sharpe Ratio From Scratch
def sharpe_ratio(data, risk_free_rate=0.0):
    # Calculate Average Daily Return
    mean_daily_return = sum(data) / len(data)
    # Calculate Standard Deviation
    s = std_dev(data)
    # Calculate Daily Sharpe Ratio
    daily_sharpe_ratio = (mean_daily_return - risk_free_rate) / s
    # Annualize Daily Sharpe Ratio
    sharpe_ratio = 252**(1/2) * daily_sharpe_ratio

    return sharpe_ratio
```

```python
def runSimulation(data, stock, options, shortEnable, riskfreerate):
    for index, row in data.iterrows():
        win = ""
        temp = stock_vs_options(row.startDate, row.endDate, row.exitStrategy, row.symbols, stock, options, shortEnable, row.percentageBuy, riskfreerate, row.dateDistance)
        data.at[index, "lossProfitStock"] = temp[0]
        data.at[index, "lossProfitOption"] = temp[1]
        data.at[index, "Sharpe_Ratio Stock"] = temp[2]
        data.at[index, "Sharpe_Ratio Option"] = temp[4]
        data.at[index, "TWR Stock"] = temp[3]
        data.at[index, "TWR Option"] = temp[5]
        if np.all((temp[0] > temp[1])): win = "Stock"
        elif np.all(temp[0] < temp[1]): win = "Option"
        elif np.all(temp[0] == temp[1]): win = "En par"
        data.at[index, "winner"] = win
    return data
```

```python
def summaryStatistics(excludedStocks = ["First", "Second", "Third"]):
    df = pd.DataFrame(columns =["percentageBuy","dateDistance", "AverageProfitStock", "AverageProfitOption", "AverageStockSharpeRatio", "AverageOptionSharpeRatio", "AverageStockTWR", "AverageOptionTWR"])
    firstpercentageBuy = 0
    firstdateDistance = 0
    firstIndex = 0
    TempProfitStock = 0
    TempProfitOption = 0
    TempStockSharpeRatio = 0
    TempOptionSharpeRatio = 0
    TempStockTWR = 0
    TempOptionTWR = 0
    i = 0
    for index, row in simulation.iterrows():
        if (excludedStocks[0] not in row.symbols) & (excludedStocks[1] not in row.symbols) & (excludedStocks[2] not in row.symbols):
            if(i == 0):
                firstpercentageBuy = row.percentageBuy
                firstdateDistance = row.dateDistance

            if(row.percentageBuy != firstpercentageBuy) | (row.dateDistance != firstdateDistance):
                df = df.append({"percentageBuy": firstpercentageBuy, "dateDistance": firstdateDistance, "AverageProfitStock": TempProfitStock/(i - firstIndex), "AverageProfitOption": TempProfitOption/(i - firstIndex),"AverageStockSharpeRatio": TempStockSharpeRatio/(i - firstIndex), "AverageOptionSharpeRatio": TempOptionSharpeRatio/(i-firstIndex), "AverageStockTWR": TempStockTWR/(i-firstIndex), "AverageOptionTWR": TempOptionTWR/(i-firstIndex)}, ignore_index = True)
                firstpercentageBuy = row.percentageBuy
                firstdateDistance = row.dateDistance
                firstIndex = i
                TempProfitStock = 0
                TempProfitOption = 0
                TempStockSharpeRatio = 0
                TempOptionSharpeRatio = 0
                TempStockTWR = 0
                TempOptionTWR = 0
                
            if(row.percentageBuy == firstpercentageBuy) & (row.dateDistance == firstdateDistance):
                TempProfitStock += simulation.at[index, "lossProfitStock"][0]
                TempProfitOption += simulation.at[index, "lossProfitOption"][0]
                TempStockSharpeRatio += simulation.at[index, "Sharpe_Ratio Stock"]
                TempOptionSharpeRatio += simulation.at[index, "Sharpe_Ratio Option"]
                TempStockTWR += simulation.at[index, "TWR Stock"]
                TempOptionTWR += simulation.at[index, "TWR Option"]
                
                    
            i += 1
    df = df.append({"percentageBuy": firstpercentageBuy, "dateDistance": firstdateDistance, "AverageProfitStock": TempProfitStock/(i - firstIndex), "AverageProfitOption": TempProfitOption/(i - firstIndex),"AverageStockSharpeRatio": TempStockSharpeRatio/(i - firstIndex), "AverageOptionSharpeRatio": TempOptionSharpeRatio/(i-firstIndex), "AverageStockTWR": TempStockTWR/(i-firstIndex), "AverageOptionTWR": TempOptionTWR/(i-firstIndex)}, ignore_index = True)
    return df
```

```python
def plot_summaryStatistics():
    data = summaryStatistics.copy()
    data=data.sort_values(by=['dateDistance', 'percentageBuy'], axis=0)
    # 3 subplots for each row
    nrows=2
    ncols=3
    

    # Create figure and axes
    bar_figure, axes = plt.subplots(nrows = nrows, ncols = ncols, figsize= (30,15))

    for row in range(0,nrows):
        for column in range(0,ncols):
            data_cropped = data.head(6)
            datay1 = data_cropped["AverageProfitOption"]
            datay2 = data_cropped["AverageProfitStock"]
            datax = data_cropped["percentageBuy"]
            axe = axes[row][column]
            axe.plot(datax, datay1, color = "Blue", label = "Option")
            axe.plot(datax, datay2, color = "Red", label = "Stock")
            axe.legend()        
            plt.sca(axe)
            axe.set_title(str(int(data.dateDistance.iloc[0])) + " days")
            plt.grid(True)
            data = data.iloc[6:]
    # Show plot
    plt.show()
```
